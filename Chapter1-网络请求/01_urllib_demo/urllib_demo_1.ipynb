```python
【urlopen 函数】
在Python3的urllib库中，所有和网络请求相关的方法，都被集到urllib.request模块下面了，以先来看下urlopen函数基本的使用：
```

```python
from urllib import request
url = 'https://www.sogou.com/'
resp = request.urlopen(url)
print(type(resp))
# 返回值是一个http.client.HTTPResponse对象，这个对象是一个类文件句柄对象。
# 有read(size)、readline、readlines以及getcode（获取当前响应状态码）等方法。
```

```python
# f.read(n) n 代表读取n个字符
print(resp.read())
```

```python
print(resp.read(10))
```

```python
print(resp.readline()) # 读取单行
```

```python
print(resp.readlines()) # 读取文件到一个列表
print(type(resp.readlines()))
```
